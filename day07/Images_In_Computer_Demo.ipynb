```
from skimage import io
import matplotlib.pyplot as plt
import numpy as np
```

```
# Reading an image
url = 'http://archive.engineering.nyu.edu/files/imagecache/img_col_8_380/pressrelease/RS46906_071A7857-scr.jpg'
img = io.imread(url)
plt.imshow(img)
print(img)
print(img.shape)
```

```
plt.figure()
plt.imshow(img)
plt.figure()
plt.imshow(img[:,:,0],'gray')
plt.title('Red')
plt.figure()
plt.imshow(img[:,:,1],'gray')
plt.title('Green')
plt.figure()
plt.imshow(img[:,:,2],'gray')
plt.title('Blue')
```

```
#Convert to grayscale
R = img[:,:,0]
G = img[:,:,1]
B = img[:,:,2]
          
im_gray = 0.2989 * R + 0.5870 * G + 0.1140 * B 
plt.figure()
plt.imshow(im_gray, cmap='gray')
# Add noise
noise = np.random.randn(im_gray.shape[0], im_gray.shape[1])
noisy = np.clip(im_gray + 20*noise, 0, 255)
plt.figure()
plt.imshow(noisy, cmap='gray')
```
